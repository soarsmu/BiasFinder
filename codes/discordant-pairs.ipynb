# Number of Discordant Pairs

```python
import pandas as pd
import numpy as np
import math
```

### Get Original Data

```python
# dir_name = "eec"
dir_name = "imdb_mutant"
dfm = pd.read_csv("../data/" + dir_name + "/male/test.csv", header=None, sep="\t", names=["label", "text", "template"])
dff = pd.read_csv("../data/" + dir_name + "/female/test.csv", header=None, sep="\t", names=["label", "text", "template"])
```

```python
dfm
```

### Get the result from prediction

```python
output_dir = "exp1_on_imdb"
result_dir = "../result/" + output_dir + "/"

rfm = result_dir + "results_data_male.txt"
rff = result_dir + "results_data_female.txt"

def read_from_txt(fpath) :    
    pred = []
    file = open(fpath)
    lines = file.readlines()
    for l in lines :
        pred.append(int(l))
    file.close()
    return pred

mpred = read_from_txt(rfm)
fpred = read_from_txt(rff)

print(len(mpred))
print(len(fpred))
```

```python
dfm["prediction"] = mpred
dff["prediction"] = fpred
```

#### Group by Template

```python
dfm["template"] = dfm["template"].astype("category")
dfm["template_id"] = dfm["template"].cat.codes

dfm
```

```python
dff["template"] = dff["template"].astype("category")
dff["template_id"] = dff["template"].cat.codes

dff
```

```python
dfm.groupby("template_id").get_group(0)
```

```python
dff.groupby("template_id").get_group(0)
```

### Implementation

```python
mgb = dfm.groupby("template_id")
fgb = dff.groupby("template_id")
```

#### Using Group By

```python
df = pd.DataFrame(columns=["male_text", "female_text", "male_prediction", "female_prediction"])
# df = df.append({"male_text" : "abc", "female_text": "def", "male_prediction" : 1, "female_prediction": 0}, ignore_index=True)
df
```

```python
# for i in range(len(mgb.size())) :
for i in range(3) :
    mdata = mgb.get_group(i)
    fdata = fgb.get_group(i)
    for mindex, mrow in mdata.iterrows():
        for findex, frow in fdata.iterrows():
            df = df.append({"male_text" : mrow["text"], "female_text": frow["text"], "male_prediction" : mrow["label"], "female_prediction": frow["label"]}, ignore_index=True)
```

```python
df["discordant"] = df["male_prediction"] != df["female_prediction"]
```

```python
df = pd.DataFrame(columns=["male_text", "female_text", "male_prediction", "female_prediction"])
# df = df.append({"male_text" : "abc", "female_text": "def", "male_prediction" : 1, "female_prediction": 0}, ignore_index=True)
df
```

```python
template_size = 20
lb = 0
while lb < len(dfm) :
# while lb < 20 :
    ub = lb + template_size
    mdata = dfm[lb:ub]
    fdata = dff[lb:ub]
    for mindex, mrow in mdata.iterrows():
        for findex, frow in fdata.iterrows():
            df = df.append({"male_text" : mrow["text"], "female_text": frow["text"], "male_prediction" : mrow["prediction"], "female_prediction": frow["prediction"]}, ignore_index=True)
    lb = ub
```
