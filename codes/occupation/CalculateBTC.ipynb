```python
templateid_to_btc = {}

for index, rows in df.iterrows() :
    templateid_to_btc[rows["template_id"]] = rows["btc"] 
```

```python
selected_templates = new_df["template_id"].drop_duplicates().reset_index(drop=True)
```

```python
selected_templates
```

```python
sum_btc = 0
for template in selected_templates :
    sum_btc += templateid_to_btc[template]
```
