```python
import pandas as pd
import numpy as np
import math
import time
```

```python
eval_dir = "gc_imdb_small"
# df = pd.read_csv("../data/" + eval_dir + "/test.csv", header=None, sep="\t", names=[0, 1, "mutant", "template", "gender", "label", "country"])
df = pd.read_csv("../data/" + eval_dir + "/test.csv", header=None, sep="\t", names=["label", "mutant", "gender","template"])
df
```

```python
def read_txt(fpath):
    pred = []
    file = open(fpath)
    lines = file.readlines()
    for l in lines :
        pred.append(int(l))
    file.close()
    
    return pred
```

```python
output_dir = "gc_imdb_small"

result_dir = "../result/" + output_dir + "/"

path = result_dir + "results_data.txt"

pred = read_txt(path)

print(len(pred))
```

```python
df["prediction"] = pred
```

### Use Groupby to Group the text by Template

```python
df["template"] = df["template"].astype("category")
df["template_id"] = df["template"].cat.codes
```

```python
gb = df.groupby("template_id")
```

```python
gb.count()
```

```python
size = gb.count()["mutant"][0]
```

```python
dsum = gb.sum()
dsum
```

```python
dsum["diff"] = abs(dsum["label"] - dsum["prediction"])

dsum
```

```python
dunfair = dsum[dsum["diff"]*2 > size]

dunfair
```
