```python
import pandas as pd
import numpy as np
import math
import time
```

```python
base_dir = "../../data/biasfinder/age/"
df = pd.read_csv(base_dir + "test.csv", header=None, sep="\t", names=["label", "mutant", "mutant2", "template", "age", "original"])
df.drop(columns=["mutant2"])
df
```

```python
def replaceSpace(text):
    return text.replace(" ", "")
df["age"] = df["age"].apply(replaceSpace)
df
```

```python
def getGroup(text) :
    if text == "old" or text == "young" :
        return "old_young"
    elif text == "older" or text == "younger" :
        return "older_younger"
    elif text == "oldest" or text == "youngest" :
        return "oldest_youngest"
    elif text == "25yearsold" or text == "65yearsold" :
        return "yearsold"
df["class"] = df["age"].apply(getGroup)
df
```

```python
def read_txt(fpath):
    pred = []
    file = open(fpath)
    lines = file.readlines()
    for l in lines :
        pred.append(int(l))
    file.close()
    
    return pred
```

```python
output_dir = "age"

result_dir = "../../result/" + output_dir + "/"

path = result_dir + "results_data.txt"

pred = read_txt(path)

print(len(pred))
```

```python
df["prediction"] = pred
```

# Create Discordant Pairs based on Age

```python
start = time.time()

identifier = "class"

mutant_text_1 = []
mutant_text_2 = []
prediction_1 = []
prediction_2 = []
identifier_1 = []
identifier_2 = []
template = []
original = []
for i in range(len(gb.size())) :
    data = gb.get_group(i)
    classes = data.groupby("class")
    for key, value in dict(iter(classes)).items() :
        
    
#         dc = data.groupby("gender")
        dc = value.groupby("age")
        already_processed = []
        for k1, v1 in dict(iter(dc)).items() :
            for k2, v2 in dict(iter(dc)).items() :
                if k1 != k2 :
                    k = k1 + "-" + k2
                    if k1 > k2 :
                        k = k2 + "-" + k1
                    if k not in already_processed :
                        already_processed.append(k)

                        for m_1, p_1, i_1, t, o in zip(v1["mutant"].values, v1["prediction"].values, v1["age"].values, v1["template"].values, v1["original"].values) :
                            for m_2, p_2, i_2 in zip(v2["mutant"].values, v2["prediction"].values, v2["age"].values) :
                                if p_1 != p_2 :
                                    mutant_text_1.append(m_1)
                                    prediction_1.append(p_1)
                                    identifier_1.append(i_1)
                                    mutant_text_2.append(m_2)
                                    prediction_2.append(p_2)
                                    identifier_2.append(i_2)
                                    template.append(t)
                                    original.append(o)

end = time.time()
print("Execution time: ", end-start)
```

```python
btc = pd.DataFrame(data={"mutant_1" : mutant_text_1, "mutant_2" : mutant_text_2, "prediction_1": prediction_1, "prediction_2" : prediction_2, "identifier_1": identifier_1, "identifier_2" : identifier_2, "template": template,  "original":original})
btc
```

```python
print("Number of BTC: ", int(len(btc)))
```

```python
btc = btc.sample(frac=1, random_state=123)

texts = []
templates = []
for index, rows in btc.iterrows():
    texts.append(rows["original"])
    texts.append(rows["mutant_1"])
    texts.append(rows["mutant_2"])
    templates.append(rows["template"])
```
